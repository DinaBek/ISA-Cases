List Apprehension

```python
txns = [15,15,23,11,42]
tax_rate = .08
```

```python
lst = []


for i in txns:
    lst.append(i*tax_rate)
print(lst)
```

```python
def get_price_with_tax(txn):
    return txn * (1 + tax_rate)
```

```python
get_price_with_tax(7)
```

```python
lst = []


for i in txns:
    lst.append(get_price_with_tax(i))
print(lst)
```

```python
final_prices = map(get_price_with_tax, txns)
final_prices = list(map(get_price_with_tax, txns))
#list(final_prices)
```

```python
final_prices
```

```python
final_prices = tuple(map(get_price_with_tax, txns))
```

```python
final_prices
```

```python
#list apprehension

final_prices = [get_price_with_tax(i) for i in txns]
print(final_prices)
```

```python
original_prices = [1.25, -9.45, 10.22, 3.78, -5.92, 1.16]
prices = [i if i > 0 else 0 for i in original_prices]
prices2 = [i for i in original_prices if i > 0]
print(prices)
print(prices2)
```

```python
def get_price(price):
    return price if price > 0 else 0

prices = [get_price(i) for i in original_prices]
prices
```

```python
list(map(get_price, original_prices))
```
