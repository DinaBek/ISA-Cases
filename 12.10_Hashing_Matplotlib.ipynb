```python
x = np.linspace(0,2,100)

fig, ax = plt.subplots()
ax.plot(x,x, label = 'linear', color = 'r')
ax.plot(x,x**2, label = 'quadric', color = 'b')
ax.plot(x,x**3, label = 'cubic', color = 'g')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple plot')
ax.legend()
ax.legend(loc = 'upper left');
```

```python
names = ['group a', 'group b', 'group c']
values = [1,10,100]

plt.figure(figsize = (9,5))

plt.subplot(2,2,1)
plt.bar(names, values)
plt.title('Bar Plot')

plt.subplot(2,2,2)
plt.scatter(names, values)
plt.title('Scatter Plot')

plt.subplot(2,2,3)
plt.plot(values, names)
plt.title('Plot Plot')

plt.suptitle('Categorical Plotting')
plt.tight_layout()
plt.show()
```

```python
def f(t):
    return np.exp(-t)* np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.figure()
plt.subplot(222)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show();
```

```python
def f(t):
    return np.exp(-t)* np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)


plt.figure()
plt.figure(figsize=(14,5))
plt.subplot(121)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')


plt.subplot(222)
plt.plot(t1, f(t1), 'bo')

plt.subplot(224)
plt.plot(t2, f(t2), 'k')
plt.tight_layout()
plt.show();
```
