Task 3. Пользователь задаёт три числа. Если среди них есть равные друг другу, то вывести YES, если нет вывести ERROR

```python
x1 = int(input("Enter the first number: "))
x2 = int(input("Enter the second number: "))
x3 = int(input("Enter the third number: "))

i = (x1,x2,x3)
if i[0] == i[1] or i[0] == i[2] or i[2] == i[1]:
    print("yes")
else:
    print("error") 
```
